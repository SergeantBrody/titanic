```python
enc_one = OneHotEncoder(cols="Embarked", drop_invariant=False, return_df=True, handle_missing='value', handle_unknown='return_nan', use_cat_names=True)
encoded_df = enc_one.fit_transform(new_df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = encoded_df.drop('Survived', axis=1)
y = encoded_df.Survived
X_train, X_test, Y_train, Y_test, = train_test_split(x, y, test_size = 0.2)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python


rfc = RandomForestClassifier(random_state=5, n_jobs=15)
rfc.fit(X_train, Y_train)
rfc.predict(X_test)
rfc.score(X_test, Y_test)
```
